```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_excel('DATOS_TEMPERATURA_HUMEDAD_SBND.xlsx', sheet_name='Sheet3')

date_strings=df['Time'].array.to_numpy()
date_format = "%m/%d/%y, %I:%M:%S %p %Z"

datetime_objects = []

for date_string in date_strings:
    datetime_objects.append(datetime.strptime(date_string,date_format))
datetime_objects = np.array(datetime_objects)
```

```python
plt.figure(figsize=(20,10),dpi=150)

print(df.keys())
for key in df.keys()[1:]:
    plt.plot(datetime_objects, df[key].array.to_numpy(), label=key)
plt.legend()
```
